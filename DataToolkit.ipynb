2.How does broadcasting work in NumPy?

Answer-2-->>>
\---> Broadcasting in NumPy means we can do operations between arrays of different shapes.
\---> It automatically adjusts the smaller array so it can match the shape of the bigger one during calculation.

\--->>> This helps us avoid writing loops and makes the code shorter and faster.

\--> For example, if we have a 2D array and we want to add a 1D array to each row, NumPy will "stretch" the smaller one to match the shape of the bigger one without copying the data.

\--->> Let's say we have a matrix of student marks, and we want to add bonus marks to each subject — broadcasting makes it super easy and quick.

21.Why is NumPy’s array slicing faster than Python’s list slicing?

Answer-21-->>>
\---> NumPy’s array slicing is faster than Python’s list slicing because NumPy arrays store data in a continuous block of memory, which allows quick access to elements.
\---> Python lists store references to objects scattered in memory, so accessing elements is slower.

\--->>> Also, NumPy slicing doesn’t create a copy of the data but returns a view of the original array, which saves time and memory.

\--> In contrast, slicing a Python list creates a new list, which takes extra time and space.

\--->> For example, when Sameer slices a big array of numbers to get a subset, NumPy does it much faster than a normal Python list.

```python
# Question:1-->>> How do you create a 2D NumPy array and calculate the sum of each row?
# Answer:1-->>>
import numpy as np

# Creating a 2D NumPy array with 3 rows and 4 columns
arr = np.array([[10, 20, 30, 40],
                [5, 15, 25, 35],
                [2, 4, 6, 8]])

# Calculating the sum of each row using axis=1
row_sums = arr.sum(axis=1)

print(row_sums)  # This will give output: [100 80 20]

# Example: Suppose Sameer has marks of students in 3 subjects in each row,
# he can use this to find total marks per student quickly.
```

```python
# Question:2-->>> Write a Pandas script to find the mean of a specific column in a DataFrame?
# Answer:2-->>>
import pandas as pd

# Creating a sample DataFrame with student names and their marks
data = {'Name': ['Sameer', 'Anjali', 'Rohit', 'Priya'],
        'Marks': [85, 90, 78, 92]}

df = pd.DataFrame(data)

# Finding the mean of the 'Marks' column
mean_marks = df['Marks'].mean()

print(mean_marks)  # This will give output: 86.25

# Example: Anjali wants to know the average marks of her class using this code.
```

```python
# Question:3-->>> Create a scatter plot using Matplotlib?
# Answer:3-->>>
import matplotlib.pyplot as plt

# Sample data: Sameer's study hours vs marks scored
study_hours = [2, 4, 6, 8, 10]
marks = [50, 60, 70, 80, 90]

# Creating scatter plot
plt.scatter(study_hours, marks, color='blue')

# Adding title and labels
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')

plt.show()

# Example: Sameer wants to visualize how his study hours relate to his marks using this scatter plot.
```

```python
# Question:4-->>> How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
# Answer:4-->>>
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a sample DataFrame with some data
data = {'Math': [85, 90, 78, 92, 88],
        'Physics': [80, 85, 75, 90, 86],
        'Chemistry': [78, 88, 80, 85, 90]}

df = pd.DataFrame(data)

# Calculating the correlation matrix
corr_matrix = df.corr()

# Plotting heatmap using Seaborn
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix Heatmap')
plt.show()

# Example: Priya wants to see how students’ scores in different subjects relate to each other,
# so she uses this heatmap to find strong or weak correlations easily.
```

```python
# Question:5-->>> Generate a bar plot using Plotly?
# Answer:5-->>>
import plotly.graph_objects as go

# Sample data: Subjects and average marks
subjects = ['Math', 'Physics', 'Chemistry', 'English']
avg_marks = [85, 78, 90, 88]

# Creating a bar plot
fig = go.Figure(data=[go.Bar(x=subjects, y=avg_marks)])

# Adding title and labels
fig.update_layout(title='Average Marks by Subject',
                  xaxis_title='Subjects',
                  yaxis_title='Average Marks')

fig.show()

# Example: Rahul uses this bar plot to compare average marks of students in different subjects.
```

```python
# Question:6-->>> Create a DataFrame and add a new column based on an existing column?
# Answer:6-->>>
import pandas as pd

# Creating a sample DataFrame with student names and marks
data = {'Name': ['Sameer', 'Anjali', 'Rahul', 'Priya'],
        'Marks': [75, 85, 90, 65]}

df = pd.DataFrame(data)

# Adding a new column 'Result' based on 'Marks' column
# If marks >= 80, then 'Pass', else 'Fail'
df['Result'] = df['Marks'].apply(lambda x: 'Pass' if x >= 80 else 'Fail')

print(df)

# Example: Sameer wants to quickly see who passed or failed based on marks using this new column.
```

```python
# Question:7-->>> Write a program to perform element-wise multiplication of two NumPy arrays?
# Answer:7-->>>
import numpy as np

# Creating two NumPy arrays
arr1 = np.array([2, 4, 6, 8])
arr2 = np.array([1, 3, 5, 7])

# Performing element-wise multiplication
result = arr1 * arr2

print(result)  # This will give output: [ 2 12 30 56]

# Example: Sameer multiplies scores of two tests for each student to analyze combined results.
```

```python
# Question:8-->>> Create a line plot with multiple lines using Matplotlib?
# Answer:8-->>>
import matplotlib.pyplot as plt

# Data for two students' monthly scores
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sameer_scores = [70, 75, 80, 85, 90]
anjali_scores = [65, 70, 75, 80, 85]

# Plotting multiple lines
plt.plot(months, sameer_scores, label='Sameer', marker='o')
plt.plot(months, anjali_scores, label='Anjali', marker='s')

# Adding title and labels
plt.title('Monthly Scores of Students')
plt.xlabel('Months')
plt.ylabel('Scores')
plt.legend()

plt.show()

# Example: Sameer and Anjali compare their monthly progress using this multi-line plot.
```

```python
# Question:9-->>> Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?
# Answer:9-->>>
import pandas as pd

# Creating a sample DataFrame with student names and marks
data = {'Name': ['Sameer', 'Anjali', 'Rohit', 'Priya'],
        'Marks': [72, 88, 65, 90]}

df = pd.DataFrame(data)

# Filtering rows where 'Marks' are greater than 80
filtered_df = df[df['Marks'] > 80]

print(filtered_df)

# Example: Priya wants to find students who scored more than 80 marks in the exam.
```

```python
# Question:10-->>> Create a histogram using Seaborn to visualize a distribution?
# Answer:10-->>>
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: marks scored by students
marks = [55, 65, 75, 80, 85, 90, 95, 70, 60, 78, 88, 92]

# Creating a histogram
sns.histplot(marks, bins=5, kde=False, color='green')

# Adding title and labels
plt.title('Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Number of Students')

plt.show()

# Example: Sameer uses this histogram to see how marks are distributed in his class.
```

```python
# Question:11-->>> Perform matrix multiplication using NumPy?
# Answer:11-->>>
import numpy as np

# Creating two 2x2 matrices
matrix1 = np.array([[1, 2],
                    [3, 4]])
matrix2 = np.array([[5, 6],
                    [7, 8]])

# Performing matrix multiplication using np.dot()
result = np.dot(matrix1, matrix2)

print(result)
# Output:
# [[19 22]
#  [43 50]]

# Example: Sameer uses matrix multiplication to solve a problem involving transformation of data.
```

```python
# Question:13-->>> Create a 3D scatter plot using Plotly?
# Answer:13-->>>
import plotly.graph_objects as go

# Sample data: marks in three subjects for different students
x = [75, 85, 65, 90]  # Math
y = [80, 70, 85, 88]  # Physics
z = [78, 90, 70, 95]  # Chemistry

# Creating 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(size=8, color='red')
)])

# Adding title and axis labels
fig.update_layout(title='3D Scatter Plot of Student Marks',
                  scene=dict(
                      xaxis_title='Math',
                      yaxis_title='Physics',
                      zaxis_title='Chemistry'
                  ))

fig.show()

# Example: Sameer uses this 3D plot to visualize how students perform in three subjects simultaneously.
```
